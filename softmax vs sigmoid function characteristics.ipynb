## Study of sigmoid,softmax,softplus and ReLU function 
##### References:
1. https://github.com/Kulbear/deep-learning-nano-foundation/wiki/ReLU-and-Softmax-Activation-Functions

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
# softmax function
def softmax(x,X):
    exp_x = [np.exp(x) for x in X]
    exp_sum = np.sum(exp_x)
    return np.exp(x)/exp_sum
```

```python
# sigmoid function
def sigmoid(x):
    return 1/(1+np.exp(-1*x))
```

```python
# generate x coordinate data
min_x = -5
max_x = 5
X = np.linspace(min_x,max_x,2*(max_x-min_x+1))
```

```python
# plot sigmoid function
plt.figure(figsize=(20,10))
sigmoid_Y = [sigmoid(x) for x in X]
plt.plot(X,sigmoid_Y)
plt.title('Sigmoid function')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# plot softmax function
plt.figure(figsize=(20,10))
softmax_Y = [softmax(x,X) for x in X]
plt.plot(X,softmax_Y)
plt.title('Softmax function')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
min_x = -20
max_x = 20
X = np.linspace(min_x,max_x,5*(max_x-min_x+1))
# plot softmax function
plt.figure(figsize=(20,10))
softmax_Y = [softmax(x,X) for x in X]
plt.plot(X,softmax_Y,label='softmax')
sigmoid_Y = [sigmoid(x) for x in X]
plt.plot(X,sigmoid_Y,label='sigmoid')
plt.title('Softmax and Sigmoid function')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```

## softplus and ReLU(Rectified Linear Unit) function

```python
# softplus function
def softplus(x):
    return np.log(1+np.exp(x))
```

```python
# ReLU function
def relu(x):
    return max(x,0)
```

```python
# plot softplus and ReLU function
min_x = -10
max_x = 10
X = np.linspace(min_x,max_x,10*(max_x-min_x+1))
# plot softmax function
plt.figure(figsize=(20,10))
softplus_Y = [softplus(x) for x in X]
plt.plot(X,softplus_Y,label='softplus')
relu_Y = [relu(x) for x in X]
plt.plot(X,relu_Y,label='ReLU')
plt.title('Softplus and ReLU function')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()
plt.show()
```
